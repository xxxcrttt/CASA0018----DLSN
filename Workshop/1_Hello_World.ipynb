# Import

Import **TensorFlow** and **numpy**

**keras**: -- Deep Learning API, running on top of the Machine Leanring platform TensorFlow.

The framework for defining a neural network as a set of sequential layers 


```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

# Provding the data




```python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

# Define and Compile the Neural Network 

using the keras **sequential** function --    
this adds layers sequentially to the model includes inputs, outputs and hidden layers. 

**input shape** to have a value of 1 -- x,   
then add 1 dense layer with a size of 1 neuron.   

**Dense** function describes a network layers where all neurons are connected to all other neurons. 

```python
model = tf.keras.Sequential()
model.add(tf.keras.Input(shape=(1,)))
model.add(tf.keras.layers.Dense(1,))
```

```python
model.summary()
```

now compile the Neural Network and specify 2 functions, a *loss* and an *optimizer*.    
**Loss**: measures the guessed answer against the known correct answer.     
-- use "Mean Squared Error(MSE)"(均方误差)   

**Optimizer**: try to optimize the loss.    
-- use "Stochastic Gradient Descent"(随机梯度下降)

```python
model.compile(optimizer="sgd", loss="mean_squared_error")
```

# Training the Neural Network 

"learns" the relationship between xs and ys in the *model.fits* call, loops and making guess,   
measuring the loss and using the optimizer to make another guess. 

```python
model.fit(xs, ys, epochs=200)
```

```python
print(model.predict([10]))
```
